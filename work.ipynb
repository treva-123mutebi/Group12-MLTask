# Univariate Visualizations:



## Histogram of Boys Enrollment:
This visualization represents the distribution of boys enrollment numbers across different years and districts. The x-axis represents the enrollment numbers, and the y-axis represents the frequency or count of districts and years falling into each enrollment bin. This histogram allows us to understand the distribution of boys enrollment and identify any patterns or concentration of enrollment numbers within specific ranges.

```python
# Calculate the maximum boys enrollment for each year
max_enrollment = df.groupby('Year')['Total Boys Enrollment'].max()

# Set up the plot
plt.figure(figsize=(10, 8))
sns.set_style("darkgrid")

# Plot the bar chart using the maximum boys enrollment
plt.bar(max_enrollment.index, max_enrollment.values, alpha=0.5, color=sns.diverging_palette(220, 20, n=1))

# Set the plot title and labels
plt.title("Year vs. Maximum Boys Enrollment")
plt.xlabel("Year")
plt.ylabel("Boys Enrolled (Maximum)")

# Save the plot as an image
plt.savefig("Images/YearvsMaxBoysEnrollment.png")

# Display the plot
plt.show()
```

# Multivariate Visualisations

## Bar Chart: Enrollment Comparison - Boys vs Girls

This bar chart compares the enrollment numbers of boys and girls over different years. Each bar represents a specific year, and the height of the bar corresponds to the total enrollment count. The chart uses two different colors, one for boys enrollment and another for girls enrollment, allowing for a visual comparison between the two genders.

The x-axis represents the years, while the y-axis represents the enrollment count. The chart helps us analyze and observe any disparities or trends in enrollment between boys and girls over time. By comparing the heights of the bars, we can identify which gender had higher or lower enrollment in each particular year.

This visualization provides a clear overview of the enrollment comparison between boys and girls and allows us to easily identify any differences or patterns in their enrollment numbers.

```python

# Group the dataset by year and calculate the sum of boys and girls enrollment
grouped_df = df.groupby('Year')['Total Boys Enrollment', 'Total Girls Enrollment'].sum()

# Set up the plot
plt.figure(figsize=(10, 8))

# Get the years and the number of bars
years = grouped_df.index
num_bars = len(years)
bar_width = 0.35

# Generate the positions of the bars
boys_bar_positions = range(num_bars)
girls_bar_positions = [x + bar_width for x in boys_bar_positions]

# Plot the boys enrollment bars
plt.bar(boys_bar_positions, grouped_df['Total Boys Enrollment'], bar_width, alpha=0.5, label='Boys Enrollment')

# Plot the girls enrollment bars
plt.bar(girls_bar_positions, grouped_df['Total Girls Enrollment'], bar_width, alpha=0.5, label='Girls Enrollment')

# Set the plot title and labels
plt.title('Enrollment Comparison: Boys vs Girls')
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.xticks(boys_bar_positions, years)
plt.legend()

# Save the plot as an image
plt.savefig('Images/EnrollmentComparison.png')

# Display the plot
plt.show()
```

## Bar Chart: Average Girls Enrollment by Type of School

This bar chart displays the average girls' enrollment for different types of schools: Day, Boarding, and Both. The x-axis represents the types of schools, while the y-axis represents the average girls' enrollment count. Each bar in the chart represents a specific type of school and its height represents the average enrollment count of girls for that category. This visualization allows for easy comparison of the average girls' enrollment across different types of schools, providing insights into potential gender disparities in enrollment between the various school types. 

```python


# Filter the dataset by "Type of School" and calculate the total girls enrollment
day_enrollment = df.loc[df['Type of school'] == 'Day', 'Total Girls Enrollment'].sum()
boarding_enrollment = df.loc[df['Type of school'] == 'Boarding', 'Total Girls Enrollment'].sum()
both_enrollment = df.loc[df['Type of school'] == 'Both', 'Total Girls Enrollment'].sum()

# Calculate the average girls enrollment for each type of school
avg_day_enrollment = day_enrollment / df.loc[df['Type of school'] == 'Day'].shape[0]
avg_boarding_enrollment = boarding_enrollment / df.loc[df['Type of school'] == 'Boarding'].shape[0]
avg_both_enrollment = both_enrollment / df.loc[df['Type of school'] == 'Both'].shape[0]

# Set up the plot
plt.figure(figsize=(10, 8))

# Plot the bar chart of "Type of School" vs average "Total Girls Enrollment"
x_labels = ['Day', 'Boarding', 'Both']
enrollment_values = [avg_day_enrollment, avg_boarding_enrollment, avg_both_enrollment]
plt.bar(x_labels, enrollment_values, alpha=0.5)

# Set the plot title and labels
plt.title('Type of School vs Average Girls Enrollment')
plt.xlabel('Type of School')
plt.ylabel('Average Girls Enrollment')

# Save the plot as an image
plt.savefig('Images/TypeOfSchoolVsAverageGirlsEnrollment.png')

# Display the plot
plt.show()
```

## Stacked Bar Chart: Boys and Girls Enrollment by Grade Level

This stacked bar chart shows the enrollment counts for boys and girls in each grade level. Each bar represents a grade level, and the heights of the bars represent the enrollment counts. The chart allows us to compare the enrollment disparity between boys and girls across different grade levels.

```python
# Select the relevant columns
grade_columns = ['Total P1', 'Total P2', 'Total P3', 'Total P4', 'Total P5', 'Total P6', 'Total P7']
boys_columns = [f'{grade} Boys Enrollment' for grade in grade_columns]
girls_columns = [f'{grade} Girls Enrollment' for grade in grade_columns]

# Calculate the average enrollment for boys and girls by grade level
boys_enrollment = df[boys_columns].mean()
girls_enrollment = df[girls_columns].mean()

# Set up the plot
plt.figure(figsize=(10, 8))

# Plot the stacked bar chart of boys and girls enrollment by grade level
plt.bar(grade_columns, boys_enrollment, label='Boys')
plt.bar(grade_columns, girls_enrollment, bottom=boys_enrollment, label='Girls')

# Set the plot title and labels
plt.title('Boys and Girls Enrollment by Grade Level')
plt.xlabel('Grade Level')
plt.ylabel('Average Enrollment Count')
plt.legend()

# Save the plot as an image
plt.savefig('Images/BoysGirlsEnrollmentByGrade.png')

# Display the plot
plt.show()
```

## Line Plot: Trend of Boys and Girls Enrollment over Time

This line plot shows the trend of boys and girls enrollment over time. The x-axis represents the years, and the y-axis represents the enrollment counts. The plot includes separate lines for boys and girls enrollment, allowing us to observe any changes or disparities in enrollment trends between the two genders over the years.

```python


# Group the dataset by year and calculate the total enrollment for boys and girls
grouped_df = df.groupby('Year')[['Total Boys Enrollment', 'Total Girls Enrollment']].sum()

# Set up the plot
plt.figure(figsize=(10, 8))

# Plot the line chart of boys and girls enrollment over time
plt.plot(grouped_df.index, grouped_df['Total Boys Enrollment'], marker='o', label='Boys')
plt.plot(grouped_df.index, grouped_df['Total Girls Enrollment'], marker='o', label='Girls')

# Set the plot title and labels
plt.title('Trend of Boys and Girls Enrollment over Time')
plt.xlabel('Year')
plt.ylabel('Enrollment Count')
plt.legend()

# Save the plot as an image
plt.savefig('Images/BoysGirlsEnrollmentTrend.png')

# Display the plot
plt.show()
```
